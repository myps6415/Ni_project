```python
import pandas as pd

df1 = pd.read_csv('/Users/john.tung/Downloads/Ni/處理過的csv/20210101_20210117_cashin.csv')
df1.head()
```

```python
df1 = df1.drop(columns=['次数'])
print(len(df1))
df1.head()
```

```python
df2 = pd.read_csv('/Users/john.tung/Downloads/Ni/處理過的csv/BB_出款20210101_20210117.csv')
df2.head()
```

```python
df2 = df2.drop(columns=['次数'])
print(len(df2))
df2.head()
```

```python
df = df1.merge(df2, how='outer')
print(len(df))
df.head()
```

```python
df.to_csv('存提款.csv', index=False)
```

```python
date_list = []
deposit_list = []
deposit_money_list = []

for date in set(df['日期']):
    date_list.append(date)
    deposit_list.append(len(set(df[(df['日期'] == date)&(df['存款金额'].notnull())]['用户名'])))
    deposit_money_list.append(sum(df[(df['日期'] == date)&(df['存款金额'].notnull())]['存款金额']))
    
deposit_pNum_money = pd.DataFrame({'日期':date_list, '存款人數':deposit_list, '存款金額':deposit_money_list})
deposit_pNum_money.head()
```

```python
deposit_pNum_money.to_csv('存款人數金額.csv', index=False)
```

```python
date_list = []
deposit_list = []
deposit_money_list = []

for date in set(df['日期']):
    date_list.append(date)
    deposit_list.append(len(set(df[(df['日期'] == date)&(df['提款金额'].notnull())]['用户名'])))
    deposit_money_list.append(abs(sum(df[(df['日期'] == date)&(df['提款金额'].notnull())]['提款金额'])))
    
deposit_pNum_getmoney = pd.DataFrame({'日期':date_list, '提款人數':deposit_list, '提款金額':deposit_money_list})
deposit_pNum_getmoney.head()
```

```python
deposit_pNum_getmoney.to_csv('提款人數金額.csv', index=False)
```

```python
date_list = []
deposit_list = []
withdrawal_list = []

for date in set(df['日期']):
    date_list.append(date)
    deposit_list.append(len(set(df[(df['日期'] == date)&(df['存款金额'].notnull())]['用户名'])))
    withdrawal_list.append(len(set(df[(df['日期'] == date)&(df['提款金额'].notnull())]['用户名'])))
    
money_people_count = pd.DataFrame({'日期':date_list, '存款人數':deposit_list, '提款人數':withdrawal_list})
money_people_count.head()
```

```python
money_people_count.to_csv('存提款人數.csv', index=False)
```

```python
df3 = pd.read_csv('/Users/john.tung/Downloads/Ni/處理過的csv/BB_as.bet20210101_20210117.csv')
print(len(df3))
df3.head()
```

```python
date_category_money = df3[['日期', '分类', '下注金额', '输赢', '结算单数']].groupby(['日期', '分类']).sum().reset_index()
date_category_money['下注金额'] = date_category_money['下注金额'].astype(int)
date_category_money['输赢'] = date_category_money['输赢'].astype(int)
date_category_money.head()
```

```python
people_count = []

for date, cate in zip(date_category_money['日期'], date_category_money['分类']):
    people_count.append(len(set(df3[(df3['日期'] == date) & (df3['分类'] == cate)]['用户名'])))

date_category_money['people_count'] = people_count
date_category_money.head()
```

```python
date_category_money.to_csv('/Users/john.tung/Downloads/Ni/處理過的csv/Fish/BB_date_category_money_pCount.csv', index=False)
```
