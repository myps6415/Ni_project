```python
import pandas as pd

df2 = pd.read_csv('/Users/myps64152002/Desktop/BB02_游戏平台_留存数据new.csv',
                  encoding="gb2312",
                  index_col=0)
print(len(df2))
print(set(df2.留存比例))
df2.head()
```

```python
df2.info()
```

```python
set(df2.留存比例)
```

```python
df2.日期 = pd.to_datetime(df2.日期, infer_datetime_format=True)
df2.最初活跃 = pd.to_datetime(df2.最初活跃, infer_datetime_format=True)
df2.head()
```

```python
df2.第N天 = df2.日期 - df2.最初活跃
df2['第N天'] = [int(str(day).split(' ')[0]) for day in df2.第N天]
df2.head()
```

```python
df2 = df2[df2['第N天'].isin([1,2,3,7,14])]
df2.head()
```

```python
df2.to_csv('/Users/myps64152002/Desktop/BB02_游戏平台_留存数据.csv', encoding='utf_8_sig', index=False)
```
