## 3. 承上題，設計一個 list 命名為 python_list，存有 1 ~ 500，印出 2 可以整除的數字

```python
python_list = list(range(1, 501))

for num in python_list:
    if num % 2 == 0:
        print(num)
```

```python
print([i for i in python_list if i % 2 == 0])
```

## 4. 承上題，印出基數 (2 無法整除)

```python
for num in python_list:
    if num % 2 != 0:
        print(num)
```

```python
print([i for i in python_list if i % 2 != 0])
```

## 6. 爬取 Yahoo 購物中心，apple 產品的資料 (含標題及售價)
(https://tw.buy.yahoo.com/search/product?p=apple)

```python
import requests
from bs4 import BeautifulSoup

url = 'https://tw.buy.yahoo.com/search/product?p=apple'

res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')

for i in soup.find_all(class_='BaseGridItem__itemInfo___3E5Bx'):
    title = i.find(class_='BaseGridItem__title___2HWui').text
    try:
        price = i.find(class_='BaseGridItem__price___31jkj').find('em').text
        print(title, price)
    except:
        price = i.find(class_='BaseGridItem__price___31jkj').text
        print(title, price)
```
